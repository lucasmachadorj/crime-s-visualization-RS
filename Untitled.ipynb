```python
import pandas as pd
```

```python
frame['mesorregiao'] = frame['mesorregiao'].apply(lambda x: x.replace("-"," "))
frame['mesorregiao'] = frame['mesorregiao'].apply(lambda x: x.replace(" ","."))
```

```python
frame['microrregiao'] = frame['microrregiao'].apply(lambda x: x.replace("-"," "))
frame['microrregiao'] = frame['microrregiao'].apply(lambda x: x.replace(" ","."))
```

```python
frame['caminho'] = frame['mes'] + "-" + frame['mesorregiao'] + "-" + frame['microrregiao'] + "-" + frame['fato'] 
```
